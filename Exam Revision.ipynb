my_array[:,1]  # Obtain all rows  (:), for the second column (1)

# General Python

### List Comprehensions

__List comprehensions__ are a __fancy way of concisely creating lists__.

```python
# Old way to create the list. This produces exactly the same output as immediately below.
my_list = []
for i in range(10):
    my_list.append(to_add[i])

# List comprehension
my_list = [to_add[i] for i in range(10)]  # Create a list, using the first 10 elements of another list
```

List comprehnsions, more generally, allow you to __apply a function on a list to create a new list__.

```python
my_list = [i**2 for i in range(10) if i%2==1]  # Creates a list of the first 9 odd numbers, squared. 
```

### Classes

__Classes__ can be thought of as 'blue prints' for creating _instances_ of objects.

```python
# Define a class
class MyClass():  
    def __init__(self, x):  # Initialization constructor
        self.x = x

    def method_of_class(y):  # Example method of the class
        return y
        
instance = MyClass(value_of_x)  # Pass a value of x, to create an instance of the class
instance.method_of_class('example')  # Use the method of the class
```

# Interactive Graphics

Functions can be wrapped using ```interact()```.

````python
# Interact
from ipywidgets import interact, interactive

def f(x):  # Example function
    return x

interact(f, x=10)  # Change passed arguments to the function, interactively
````

Unlike ```interact()```, ```interactive()``` returns a Widget _instance_ rather than immediately displaying the widget.

````python
# Interactive
def f(a, b, c):  # Example function, for the widget
    return a, b, c

w = interactive(f, a=10, b=True, c=['apples', 'oranges'])  # Instantiate the interactive widget

display(w)  # Display the widget
````
