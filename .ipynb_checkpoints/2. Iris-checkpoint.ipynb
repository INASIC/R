# Feature Distributions 

Compare how the distributions of features vary for __different species__.

````r
# Plot and compare matrix of histograms for the features of each of the species of iris
par(mfrow = c(2, 2))  # Dimensions for how the plots are displayed (2x2 matrix)

# Sepal Lengths
hist(iris$Sepal.Length[iris$Species=="setosa"], col='red', xlim=c(4,9), ylim=c(0,25), main='Sepal Length (cm)', xlab='')
hist(iris$Sepal.Length[iris$Species=="versicolor"], col='green', add=T)
hist(iris$Sepal.Length[iris$Species=="virginica"], col='blue', add=T)

# Sepal Widths
hist(iris$Sepal.Width[iris$Species=="setosa"], col='red', xlim=c(2,7), ylim=c(0,30), main='Sepal Width (cm)', ylab ='', xlab='')
hist(iris$Sepal.Width[iris$Species=="versicolor"], col='green', add=T)
hist(iris$Sepal.Width[iris$Species=="virginica"], col='blue', add=T)
legend("topright", c("Setosa", "Versicolor", "Virginica"), col=c("red", "green", "blue"), lwd=5)

# Petal Lengths
hist(iris$Petal.Length[iris$Species=="setosa"], col='red', xlim=c(0,8), ylim=c(0,25), main='Petal Length (cm)',xlab='')
hist(iris$Petal.Length[iris$Species=="versicolor"], col='green', add=T)
hist(iris$Petal.Length[iris$Species=="virginica"], col='blue', add=T)

# Petal Widths
hist(iris$Petal.Width[iris$Species=="setosa"], col='red', xlim=c(0,3), ylim=c(0,40), main='Petal Width (cm)', ylab='',xlab='')
hist(iris$Petal.Width[iris$Species=="versicolor"], col='green', add=T)
hist(iris$Petal.Width[iris$Species=="virginica"], col='blue', add=T)
````

````r
par(mfrow = c(2, 2))  # Dimensions for how the plots are displayed
colBoxPlot <- c("red","green","blue")  # Colours used for the boxplot
boxplot(data=iris, Sepal.Length ~ Species, main="Sepal Length (cm)", col=colBoxPlot)  # Create boxplot for sepal length for species
boxplot(data=iris, Sepal.Width ~ Species, main="Sepal Width (cm)", col=colBoxPlot)
boxplot(data=iris, Petal.Length ~ Species, main="Petal Length (cm)", col=colBoxPlot)
boxplot(data=iris, Petal.Width ~ Species, main="Petal Width (cm)", col=colBoxPlot)
````

## Pearson Correlation Matrix

$$-1 \leq r = \frac{\sigma_{xy}}{\sqrt{\sigma_{xx} \sigma{yy}}} = \frac{\sum_{i=1}^{n}(x_{i}-\bar{x})(y_{i}-\bar{y})}{\sqrt{\sum_{i=1}^{n}(x_{i}-\bar{x})^{2}\sum_{i=1}^{n}(y_{i}-\bar{y})^{2}}} \leq 1$$

````r
# First, create an empty Pearson correlation coefficient matrix
features = iris[1:4]  # Attribute variables of the iris dataset (i.e. excluding Species) 
corr_matrix = matrix(, nrow=length(features), ncol=length(features))  # Create empty Pearson correlation matrix
corr_matrix
````

````r
# Function that converts a column dataframe into a vector
convert_df_to_numeric_vec = function(df){
    X = c()
    for (i in seq(1:nrow(df))){
      X = c(X, df[i,])
    }
    return(X)
}
````

````r
# Example of using this function on the first feature of the iris dataset: sepal length.
example = convert_df_to_numeric_vec(iris[1])
example
````

````r
mean(example)
````

````r
X = features[1]
Y = features[2]

X = convert_df_to_numeric_vec(X)
Y = convert_df_to_numeric_vec(Y)  
````

````r
sum((X-mean(X)) * (Y-mean(Y)))
````

````r
sum((X-mean(X)) * (Y-mean(Y))) / (sqrt(sum((X-mean(X))^2) * sum((Y-mean(Y))^2)))
````

````r
# Compute elements of Pearson correlation coefficient matrix
for (i in length(features)){  # Loop over each variable's correlation pair 
    X = features[i] 
    for (j in length(features)){
        Y = features[j]
        
        # Convert X and Y into vectors, to enable simpler math operations
        X = convert_df_to_numeric_vec(X)
        Y = convert_df_to_numeric_vec(Y)  
 
        # Compute Pearson correlation coefficient for the current pair of variables
        r = sum((X-mean(X)) * (Y-mean(Y))) / (sqrt(sum((X-mean(X))^2) * sum((Y-mean(Y))^2)))
    
        cat('r =',r, ' ... (i =',i,', j =', j,')')  # Print results
#         corr_matrix[i][j] = p  # Insert Pearson correlation coefficient into empty matrix
    }
}
````
