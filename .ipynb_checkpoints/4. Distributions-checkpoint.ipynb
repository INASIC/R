# Statistics for Data Analysis: R Exercises 4

R provides functions to handle commonly used random variables. For every distribution, there are four functions. There is a root name, for example, the root name for the normal distribution is norm. This root is preﬁxed by one of the letters:

* p for “probability”, the (cumulative) distribution function. 
* q for “quantile”, the inverse of the distribution function (i.e. for a given p, this gives you the value of x such that F(x) = p). 
* d for “density”, the density function. 
* r for “random”, this generates a sample (of the chosen size) from the random variable.

For example, for the normal distribution, these functions are pnorm, qnorm, dnorm, and rnorm. For the binomial distribution, these functions are pbinom, qbinom, dbinom, and rbinom. You can ﬁnd a (probably not complete) list of the available random variables in the R slides. Note that you need to input the speciﬁc parameters that determine each random variables. For example, the the normal distribution needs the mean and the standard deviation to be speciﬁed:


``` R 
# Define a grid of x values 
x<-seq(-4,4,len=100) # equispaced grid of 100 point between -4 and 4.

# standard normal density function in this range 
y<-dnorm(x,mean=0,sd=1)
plot(x,y,type='l')
```

__Try out diﬀerent values for the mean and variance to see how the density function changes__. Note that you may need to deﬁne x in a diﬀerent interval!


````r
x<-seq(-4,4,len=100) 
y<-dnorm(x,mean=0,sd=1)  # generates a normal distribution of mean 0, std 1
y2=dnorm(x,mean=1,sd=1)
y3=dnorm(x,mean=-1,sd=1)

# Plot the variables
plot(x,y,type='l', main='Normal Distributions')  # Plot y against x, with lines
points(x,y2,type='l', col='red')  # points() overlays plots over plot()
points(x,y3,type='l', col='blue')
points(x,dnorm(x,mean=0,sd=2), col='green', type='l')  # Generate y values within the plot() function

# Create a legend
legend(1.8, 0.4, legend=c("m=0, std=0", "m=1, std=0", "m=-1, std=0","m=0, std=2"), col=c("black", "red", "blue", "green"), lty=1, cex=0.8)
````

## Exercise 1 

Wires manufactured for use in a computer system are speciﬁed to have resistances between 0.12 and 0.14 ohms. The actual measured resistances produced by a company have a normal distribution with mean 0.13 ohms and standard deviation 0.005 ohm.

* __What is the probability that a randomly selected wire will meet the speciﬁcations__? 
* __What is the symmetric interval around the mean such that the probability of a randomly selected wire to belong to this interval is 0.95__? 

```R
help(pnorm)
q = c(quantile1,quantile3)  # Vector of quantiles
pnorm(q, mean = 0, sd = 1, lower.tail = TRUE, log.p = FALSE)  # Generates the cumulative probability
```

First, let us visualize the distribution of wire resistances as described in the question. 

Take note, that we are seeking the area under the curve that is between $ 0.12 \leq x \leq 0.14$.

````r
help(pnorm)
````

````r
x = seq(0.11,0.15,len=200) 
plot(x,dnorm(x,0.13,0.005), type='l', main='Resistances of Manufactured Wires', xlab='Resistance (ohms)', ylab='')
abline(v=0.12, col = "grey"); abline(v=0.14, col='grey')  # Add lines that show limits of the area we seek
````

We are seeking the area under the curve between the grey lines.

Alternatively, (and perhaps more suitably) we can determine the proportion of the y-values in this range using the ```pnorm()``` distribution function:

````r
# We have a normal distribution,
plot(x, pnorm(x,mean=0.13,sd=0.005), type='l', ylab='pnorm(x)', xlab='Resistance (ohms)', main='Probability Function of Wire Resistances')  # Generates the cumulative probability
abline(v=0.12, col = "grey"); abline(v=0.14, col='grey')  # Add lines that show limits of the area we seek
````

We can easily take the area under this distribution curve by subtracting the value of the curve at 0.14, from the value of the curve at 0.12. i.e. 

$$ p(\text{randomly selected wire meets specifications}) = p(0.12 \leq \text{wire resistance} \leq 0.14) = \text{pnorm}(0.14) - \text{pnorm}(0.12) .$$ 

Using code, we calculate this probability as:

````r
pnorm(0.14,mean=0.13,sd=0.005)-pnorm(0.12,mean=0.13,sd=0.005)
````

#### __What is the symmetric interval around the mean such that the probability of a randomly selected wire to belong to this interval is 0.95__? 

The ```qnorm()``` R function returns the __quantile function__ and specifies the value of the random variable (in this case wire resistance) such that the probability of the variable being less than or equal to that value equals the given probability. It is simply the __inverse of the cumulative distribution function__.

````r
help(qnorm)
````

````r
plot(x, qnorm(x,0.13,0.005), main='Quantile Function of Wire Resistance', xlab='Resistance (ohms)', type='l')
````

Note, that we can determine the height along the graph (say of our upper at lower limits $x_{l}=0.025$ and $x_{u}=0.975$) above using the following function:

````r
# Obtains limits of interval of interest, such that probability is 0.95
lower_limit = qnorm(0.025,0.13,0.005)  
upper_limit = qnorm(0.975,0.13,0.005)

lower_limit
upper_limit
````

We have thus determined the interval around the mean such that the probability of a randomly selected wire to belong to this interval is 0.95. 

We can visualize this interval on its density function:

````r
x = seq(0.11,0.15,len=200) 
plot(x,dnorm(x,0.13,0.005), type='l', main='Interval such that Probability=0.95', xlab='Resistance (ohms)', ylab='')
abline(v=lower_limit, col = "blue"); abline(v=upper_limit, col='red')  # Add lines that show limits of the area we seek
legend(0.135, 75, legend=c("lower_limit=0.1202", "upper_limit=0.1398"), col=c("blue", "red"), lty=1, cex=0.8)
````

We can check that it is symmetric, by checking if the mean about the interval of interest is equal to the population mean of the measured wire resistances.

````r
(qnorm(0.975,0.13,0.005) + qnorm(0.025,0.13,0.005)) / 2
````

Which confirms that it's symmetric.

Furthermore, we can double-check that our interval limits are correct by plugging these values back into the cumulative distribution function (the inverse of the ```qnorm```, which we used to determine these limits) and subtracting their heights along the the cumulative distribtution function:

````r
pnorm(upper_limit,mean=0.13,sd=0.005) - pnorm(lower_limit,mean=0.13,sd=0.005)
````

Hence, the probability that a randomly selected wire's resistance is within this symmetric interval is indeed equal to 0.95 as required.

$$p(0.1202 \leq \text{wire_resistance} \leq 0.1398) = 0.95$$ 

# Exercise 2 

The probability that a patient recovers from a severe stomach disease is 0.8. Suppose that 20 people have contracted this disease.

* Plot the density function of the random variable Y = number of people that recover from the disease. 
* What is the probability that exactly 14 people survive? 
* What is the probability that at least 10 people survive?


Here we have a binomial distributed random variable, as we have a binary random variable (0=failed to recover, 1=recovered from the disease).

````r
help(dbinom)
````

#### Plot the density function of the random variable Y = number of people that recover from the disease.

````r
# Plot the density function of the binomial random variable with n=20, p=0.8
x = 1:20  # 20 people
plot(x, dbinom(x,20,0.8), type='h', main='Stomach Disease', ylab='Probability that they recover', xlab='Number of people')  
````

### What is the probability that exactly 14 people survive?

````r
help(dbinom)
````

````r
dbinom(14,20,0.8)  # Density function of the binomial random variable
````

This value is simply the __height of the density graph above at x=14__.

$$\therefore p(\text{number of people that survived}=14) = 0.109$$

### What is the probability that at least 10 people survive?

We can determine this by using the __cumulative frequency__. We are seeking the range of y values in the interval that are at least 10, in other words: $x \geq 10 = x > 9$.

````r
# Plot the density function of the binomial random variable with n=20, p=0.8
x = 1:20  # 20 people
plot(x, pbinom(x,20,0.8), type='h', main='Cumulative Distribution Plot - Stomach Disease', ylab='Cumulative probability that they recover', xlab='Number of people')  
abline(h=pbinom(9,20,0.8), col = "blue", lty=2) # Add lines that show limits of the area we seek
````

We are seeking the total amount of y-area above the blue line. Notice that this is almost the entire graph.

Since the total cumulative frequency density is summed to 1, we can determine the probability that at least 10 people survived by subtracting from 1 the height of the cumulative frequency plot at x=9.

$$p(x_{\text{survived}} \geq 10) = 1 - p_{cumulative}(x_{\text{survived}} = 9)$$ 

````r
1-pbinom(9,20,0.8)   # Use the cumulative distribution function for the binomial random variable
````

$$\therefore p({\text{number of people that survive}} \geq 10) = 0.9994$$ 

# Exercise 3 

The number of ﬁshes caught by a ﬁsherman is distributed as a __Poisson__ random variable and the average number of ﬁshes caught in an hour is 10.

* Plot the density function of the number of ﬁshes caught in a hour. 
* What is the probability that exactly 4 ﬁshes are caught in a hour? 
* What is the probability that at least 4 ﬁshes are caught in a hour? 
* What is the minimum number of minutes in which the ﬁsherman has probability 0.9 of getting at least 4 ﬁshes?


### Plot the density function of the number of fishes caught in a hour.

The random variable (number of fishes caught by a fisherman) follows a __Poisson distribution__.

A random variable that follows a __Poisson distribution__ has a _mean that is equal to its variance_:

$$\mu = \sigma^2 = \lambda$$

Its __distribution__ is a function of its mean and random obervations, x, that only assumes positive discrete values :

$$f(x) = \frac{\mu^{x}e^{-\mu}}{x!}, \quad \text{for x = 0,1,2, ...}$$

Here we will visualize this distribution, for the case of the number of fishes caught in an hour, using the `dpois()` function:

````r
# Y is a Binomial random variable with n=20 and p=0.8
x = 0:20
lambda = 10
f = dpois(x,lambda)
plot(x,f,type='h', xlab='Number of fish caught per hour', ylab='Relative Frequency', main='Poisson Distribution - Rate of Fish caught by a Fisherman')
legend(15, 0.12, legend=c("lambda=10"), col=c("black"), lty=1, cex=0.8)
````

#### What is the probability that exactly 4 fishes are caught in a hour?

````r
dpois(4,lambda)
````

#### What is the probability that at least 4 fishes are caught in a hour?

````r
1-ppois(3,lambda)  # Subtract from 1 the cumulative relative frequency of the poisson at x=3. 
````

#### What is the minimum number of minutes in which the fisherman has probability 0.9 of getting at least 4 fishes?

````r
minutes = 1:60  # Guess a reasonable amount of minutes

# Calculate probability for all the considered number of minutes
pb = rep(NA,length(minutes))
for (k in 1:length(minutes)){  # Loop over each index (each possible number of minutes)
pb[k] = 1-ppois(3,10*(minutes[k]/60))
}

th = which(pb>=0.9)[1]  # Minimum number of minutes

# Visualize result 
plot(minutes,pb,type='l', main=(minutes[th]))
abline(v=minutes[th])  # Add vertical line where the smallest one, for which the probability is at least 0.9
````
