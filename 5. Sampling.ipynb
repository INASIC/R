What do you see? Repeat the exercise in the case the population is simulated from a normal distribution with zero mean and variance 1. Which diﬀerences can you see? 

````r
set.seed(1000)  # Ensure randomness is reproducible
````

````r
# Compare histogram of n=1000 random variable 0<=x<=1 to uniform pdf
X<-runif(1000,0,1)
hist(X,probability = TRUE)
x<-seq(0,1,len=100)
points(x,dunif(x,0,1),type='l',lwd='2')
````

Note that the histogram is normalized, such that the area is 1. This can be quickly estimated, by observing that the area of the histogram, which is a rectangle, is approximately 1.0 * 1.0 = 1. Hence we can visually see that the histogram is normalized.

````r
# Compute mean and variance of our generated sample
n = 1000  # Sample size
X = runif(n,0,1)  # Generate sample
mean(X)  # Mean of sample
((n-1)/n) * var(X)  # Unbiased estimator of variance
````

We should expect, due to the central limit theorem, that that as $n \rightarrow \infty$ the sample mean would approach $\mu \rightarrow 0.5$ and the sample variance $\sigma^{2} \rightarrow \frac{1}{12}$. With a sample size of $n=1000$, this already looks like the case. 

Let us confirm it by sampling extremely large values of n, and see if it converges to these values - by the law of large numbers (central limit theorem).

````r
# Compute the means and variances for lots of values of n
means = c()
vars = c()
ns = c()
for (n in 1:1000) {
    n=n*10
    ns[n/10] = n  
    X = runif(n,0,1)  # Generate sample
    means[n/10]=mean(X)  # Mean of sample
    vars[n/10]= var(X)  # Sample variance
}

# Visualize the results
plot(ns, means, main='Mean - Central Limit Theorem', xlab='n', ylab='Mean', type='l')
abline(h=0.5, col='red', lwd=3) 
plot(ns, vars, main='Variance - Central Limit Theorem', xlab='n', ylab='Variance', type='l')
abline(h=1/12, col='red', lwd=3)  # Overlay true population variance
````

## Exercise 2 

The probability for a patient to recover after receiveing a drug is 0.5. 200 experiments are performed with 50 patients each. Simulate the outcome of these experiments using the rbinom function and draw a histogram of the sample proportions from the experiments.

````r
n=50  # Number of patients
p=0.5  # Probability of recovery after receiving drug
X=rbinom(200,n,p)  # Sample 200 times from random binomial variable that predict recovery.  
sample_prop<-X/n
hist(sample_prop,probability = TRUE,main=paste("Chance of recovery after taking drug: p =",p))
z<-seq(0,1,len=1000)
points(z,dnorm(z,mean=p,sd=sqrt(p*(1-p)/n)),type='l', col='blue')
````

Repeat the procedure when the probability of success of the drug is __0.95__ and comment on the diﬀerence.

````r
n=50  # Number of patients
p=0.95  # Probability of recovery after receiving drug
X=rbinom(200,n,p)  # Sample 200 times from random binomial variable that predict recovery.  
sample_prop<-X/n
hist(sample_prop,probability = TRUE,main=paste("Chance of recovery after taking drug: p =",p))
z<-seq(0,1,len=1000)
points(z,dnorm(z,mean=p,sd=sqrt(p*(1-p)/n)),type='l', col='blue')
````

Here we observe that the density function has been shifted to the right. __Let us overlay the plots__:

````r
n=50  # Number of patients
p=0.95  # Probability of recovery after receiving drug
X=rbinom(200,n,p)  # Sample 200 times from random binomial variable that predict recovery.  
sample_prop<-X/n
z<-seq(0,1,len=1000)
plot(z,dnorm(z,mean=p,sd=sqrt(p*(1-p)/n)),type='l', col='red', ylab='Density')

p=0.5  # Probability of recovery after receiving drug
X=rbinom(200,n,p)  # Sample 200 times from random binomial variable that predict recovery.  
sample_prop<-X/n
z<-seq(0,1,len=1000)
points(z,dnorm(z,mean=p,sd=sqrt(p*(1-p)/n)),type='l', col='blue')
````

Indeed, we observe that when the probability of recovery after taking the drug is p=0.95, we observe a sharper and higher peak that is more distributed towards z=1. On the other hand, when p=0.5, the distribution is shorter and wider. This because their areas are normalized and must integrate to 1.
